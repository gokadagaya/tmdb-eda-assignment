```python
import requests
import pandas as pd
import sqlite3
from google.colab import userdata

# Get API key from Colab secrets
api_key = userdata.get('TMDB_KEY')

url = "https://api.themoviedb.org/3/discover/movie"

params = {
    "api_key": api_key
}

response = requests.get(url, params=params)

print(response.json())
```

At First , I stored a API key and import using user data. next, i have given the websote url to fetch the movie endpoint and print the response in json format

```python
data=response.json()
movies=data['results']
df=pd.DataFrame(movies)
print(df["genre_ids"].explode().value_counts())
```

Next i converted json response into data frame and also i printed the number of movies per genre.


```python
df = df[[
    "id", "title", "release_date",
    "popularity", "vote_average",
    "vote_count"
]]

```

Here i selected the useful columns

```python
conn = sqlite3.connect("movies.db")
df.to_sql("movies", conn, if_exists="replace", index=False)

print("Data saved to SQLite successfully!")
```

save it to a SQLite database table called movies.

```python
df = pd.read_sql("SELECT * FROM movies", conn)
print(df.head())
print("Describe..........\n")
print(df.describe())
print("is_null.........\n")
print(df.isnull().sum())
```
